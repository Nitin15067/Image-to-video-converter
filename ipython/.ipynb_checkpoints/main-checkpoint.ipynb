```python
import cv2
import numpy as np
import glob
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import random
```

```python
def get_text_window_background(window_size, background_color):
    
    image = Image.new('RGB', window_size, color = background_color)
    return image

def min_to_sec(minutes):
    return int(minutes * 60)

def get_fonts():
    font_list = []
    for filename in glob.glob('fonts/bin/*.ttf'):
        font_list.append(filename)
    return font_list
    
def get_rand_font(fonts):
    rand_font = random.choice(fonts)
    print rand_font
    return rand_font
    
    
def get_rand_font_size(a, b, c):
    return random.randrange(a, b, c)
    
# def rand_size():
    
    
# def get_text_location():
    
# def get_text_on_image_location():
    
# def get_text_color():
    
# def get_text_on_image_color():
    
# def get_text_window_background_color():
    

    
```

```python
data = {}
transitions = ["blur", "transition_1", "transition_2"]
fonts = get_fonts()
texts_array = []
text_on_images_array = []
image_array = []
font_size_min = 25
font_size_max = 40
font_size_rand_increment = 1
window_size = (1200, 675)


key = 0
file_count = 0
intro = 0
choice = 0
intro_text_temp = 0
    
for filename in glob.glob('images/*.jpg'):
     file_count = file_count + 1

print "Welcome to the Video Editor"

while(1):
    print "\n"
    print "++++++++++++++++++++++++++++++++++++++++++"
    print "We detect " + str(file_count) + " images"
    print "++++++++++++++++++++++++++++++++++++++++++"
    print "1. Enter an introductory image text"
    print "2. Enter image texts"
    print "3. Exit"
    choice = input("Enter your choice : ")

    if choice == 1:
        if intro_text_temp != 1:
            intro_text = raw_input("Enter Intro text : ")
            texts_array.append(intro_text)
            image_array.append(None)
            text_on_images_array.append(None)
            intro_text_temp = 1
        else: print "Intro already added please move to image texts"
    elif choice == 2:            
        for filename in glob.glob('images/*.jpg'):
            
            image_array.append(Image.open(filename))
            t_b_i = 0
            t_o_i = 0
            choice_1 = 0
            
            while choice_1 !=3:
                print "\n"
                print "++++++++++++++++++++++++++++++++++++++++"
                print "Image selected : " + str(filename)
                print "++++++++++++++++++++++++++++++++++++++++"
                print "1. Enter text before image " 
                print "2. Enter text on image "
                print "3. Next Image "
                choice_1 = input("Enter your choice : ")
                
                if choice_1 == 1:
                    if t_b_i !=1 :
                        text_before_image = raw_input()
                        texts_array.append(text_before_image)
                        t_b_i = 1
                    else: print "text already added move to next image"
                elif choice_1 == 2:
                    if t_o_i !=1:
                        
                        text_on_image = raw_input()
                        text_on_images_array.append(text_on_image)
                        t_o_i = 1
                    else: print "text already added move to next image"
                elif choice_1 == 3: 
                    if t_b_i == 0 :
                        texts_array.append(None)
                    if t_o_i == 0:
                        text_on_images_array.append(None)
                else: print "Wrong Input. Please enter again."
        break     
    else: print "Wrong Input. Please enter again. \n"
   
```

```python
for c in range(len(image_array)):
    
    text = texts_array[c]
    text_color = "#ffffff"
    text_image_background_color = "#000000"
    text_font_size = get_rand_font_size(font_size_min, font_size_max, font_size_rand_increment)
    text_location = (50, 50)
    img = image_array[c]
    text_on_image_color = "#ffffff"
    text_on_image_font_size = get_rand_font_size(font_size_min, font_size_max, font_size_rand_increment)
    text_on_image_location = (50, 50)
    text_on_image = text_on_images_array[c]

    
    # initializing data dictionary
    data[key] = {}
    
    # initializing window size
    data[key]['window_size'] = window_size
    
    
    # initializing text settings
    if text == None:
        data[key]['text_window'] = 0
    else :
        data[key]['text_window'] = 1
        data[key]['text'] = text
        data[key]['text_color'] = text_color
        data[key]['text_font'] = get_rand_font(fonts)
        data[key]['text_font_size'] = text_font_size
        data[key]['text_image_background_color'] = text_image_background_color
        data[key]['text_location'] = text_location
        data[key]['text_timing'] = 2
        data[key]['text_transition'] = transitions[0]
        data[key]['text_transition_timing'] = 0.2

    # initializing image settings
    if img == None:
        data[key]['img_window'] = 0
    else :
        data[key]['img_window'] = 1
        data[key]['img'] = img
        data[key]['img_timing'] = 4
        data[key]['img_transition'] = transitions[0]
        data[key]['img_transition_timing'] = 0.8

    # initializing text on image settings
    if text_on_image == None:
        data[key]['text_on_image_window'] = 0
    else :
        data[key]['text_on_image_window'] = 1
        data[key]['text_on_image'] = text_on_image
        data[key]['text_on_image_color'] = text_on_image_color
        data[key]['text_on_image_font_size'] = text_on_image_font_size
        data[key]['text_on_image_font'] = get_rand_font(fonts)
        data[key]['text_on_image_location'] = text_on_image_location
        data[key]['text_on_image_timing'] = 4
        data[key]['text_on_image_transition'] = transitions[0]
        data[key]['text_on_image_transition_timing'] = 1

    key = key + 1
    

out = cv2.VideoWriter('result/project.avi',cv2.VideoWriter_fourcc(*'DIVX'), 60, window_size)
```

```python
for d in data :
    
    # window settings
    window_size = data[d]['window_size']
    text_window = data[d]['text_window']
    text_on_image_window = data[d]['text_on_image_window']
    img_window = data[d]['img_window']
    
    # Initializing timing settings
    text_timing_in_sec = 0
    text_transition_timing_in_sec = 0
    img_timing_in_sec = 0
    img_transition_timing_in_sec = 0
    text_on_image_timing_in_sec = 0
    text_on_image_transition_timing_in_sec = 0

    # If text window is present
    if text_window == 1:
        
        # text settings
        text = data[d]['text']
        text_color = data[d]['text_color']
        text_image_background_color = data[d]['text_image_background_color']
        text_font = data[d]['text_font']
        text_font_size = data[d]['text_font_size']
        text_location = data[d]['text_location']
        text_timing = data[d]['text_timing']
        text_transition = data[d]['text_transition']
        text_transition_timing = data[d]['text_transition_timing']
        
        text_timing_in_sec = min_to_sec(text_timing)
        text_transition_timing_in_sec = min_to_sec(text_transition_timing)

        # getting text image
        text_window_background = get_text_window_background(window_size, text_image_background_color)  
        text_window_background_image = ImageDraw.Draw(text_window_background)
    
        # creating text on text image
        font = ImageFont.truetype(text_font, text_font_size)
        text_window_background_image.text(text_location, text, fill=text_color, font=font)
    
        text_window_background.save('result/text.jpg')
    
        # reading our text image
        text_img = cv2.imread('result/text.jpg')
    
        
    # if image window is present
    if img_window == 1:
        
        # image settings
        img = data[d]['img']
        img_timing = data[d]['img_timing']
        img_transition = data[d]['img_transition']
        img_transition_timing = data[d]['img_transition_timing']
        img_timing_in_sec = min_to_sec(img_timing)
        img_transition_timing_in_sec = min_to_sec(img_transition_timing)

        # if text on image is present
        if text_on_image_window == 1:
            
            # text on image settings
            text_on_image_window = data[d]['text_on_image_window']
            text_on_image = data[d]['text_on_image']
            text_on_image_font = data[d]['text_on_image_font']
            text_on_image_font_size = data[d]['text_on_image_font_size']
            text_on_image_location = data[d]['text_on_image_location']
            text_on_image_color = data[d]['text_on_image_color']
            text_on_image_timing = data[d]['text_on_image_timing']
            text_on_image_transition = data[d]['text_on_image_transition']
            text_on_image_transition_timing = data[d]['text_on_image_transition_timing']    

            text_on_image_timing_in_sec = min_to_sec(text_on_image_timing)
            text_on_image_transition_timing_in_sec = min_to_sec(text_on_image_transition_timing)

            #getting image 
            img_temp = ImageDraw.Draw(img)

            #creating text on image
            font = ImageFont.truetype(text_on_image_font, text_on_image_font_size)
            img_temp.text(text_on_image_location, text_on_image, fill=text_on_image_color, font=font)

            
        img.save('result/image.jpg')
        img = cv2.imread('result/image.jpg')

    # alogrithm for text slide
    for j in range (text_timing_in_sec):

        if j == 0 :
            B = text_transition_timing_in_sec
        if j == text_timing_in_sec - text_transition_timing_in_sec - 1 :
            B = 1

        text_image_final = text_img

        if(j < text_transition_timing_in_sec):
            text_img_final = cv2.blur(text_img,(B,B))
            B = B - 1

        if(j > text_timing_in_sec - text_transition_timing_in_sec):
            text_img_final = cv2.blur(text_img,(B,B))
            B = B + 1

        out.write(text_img_final) 

    
    # algorithm for image slide
    for k in range (img_timing_in_sec) :
        if k == 0 :
            B = img_transition_timing_in_sec
        if k == img_timing_in_sec - img_transition_timing_in_sec -1 :
            B = 1

        img_final = img

        if k < img_transition_timing_in_sec :
            img_final = cv2.blur(img,(B,B))
            B = B - 1

        if k > img_timing_in_sec - img_transition_timing_in_sec :
            img_final = cv2.blur(img,(B,B))
            B = B + 1
            
        out.write(img_final)
    
out.release()
print "yes"
        
```
